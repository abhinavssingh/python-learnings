## Types of Functions in Python

### 1. Built-in Functions
```python
print("Hello, World!")
length = len([1, 2, 3]) #output 3
```

### 2. User-defined Functions
```python
def greet(name):
    return f"Hello, {name}!"
```

### 3. Anonymous (Lambda) Functions
```python
add = lambda x, y: x + y
```

### 4. Recursive Functions
```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
```

### 5. Higher-order Functions
```python
def apply_func(func, value):
    return func(value)
```

### 6. Generator Functions
```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
```

## Types of Arguments in Python Functions

### 1. Positional Arguments
```python
def add(a, b):
    return a + b
```

### 2. Keyword Arguments
```python
def greet(name, message):
    return f"{message}, {name}!"
```

### 3. Default Arguments
```python
def power(base, exponent=2):
    return base ** exponent
```

### 4. Variable-length Arguments
```python
def sum_all(*args):
    return sum(args)

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

## Functional Programming Helpers (`map`, `filter`, `reduce`)

### map()
```python
numbers = [1, 2, 3]
squared = list(map(lambda x: x ** 2, numbers)) #output [1,4,9]
```

### filter()
```python
even = list(filter(lambda x: x % 2 == 0, [1, 2, 3, 4])) #output [2,4]
```

### reduce()
```python
from functools import reduce
product = reduce(lambda x, y: x * y, [1, 2, 3, 4]) #output 24
```

## Object-Oriented Programming (OOP) Concepts in Python

### 1. Classes and Objects
```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print(f"{self.name} says woof!")
```

### 2. Inheritance
```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Cat(Animal):
    def speak(self):
        print("Meow")
```

### 3. Encapsulation
```python
class Person:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name
```

### 4. Polymorphism
```python
class Bird:
    def sound(self):
        print("Some sound")

class Sparrow(Bird):
    def sound(self):
        print("Chirp")

class Parrot(Bird):
    def sound(self):
        print("Squawk")
```

### 5. Abstraction
```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

## Error Handling in Python

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")
```
```python
try:
    value = int("abc")
except (ValueError, TypeError):
    print("Invalid input!")
```
```python
try:
    num = int("5")
except ValueError:
    print("Invalid")
else:
    print(num)
finally:
    print("Done")
```
```python
def divide(a, b):
    if b == 0:
        raise ValueError("Denominator cannot be zero.")
    return a / b
```


````python
# example of single class and function to connect to database, execute query, and calculate salary growth
import sqlite3

db_name = 'employee_financial_insights.db'
select_query = "SELECT first_name, last_name, email, department, base_salary_monthly,query FROM employees"
growth_rate = 0.05
no_of_years = 5

class SqlConnection:
    def __init__(self, db_name, query):
        self.db_name = db_name
        self.query = query
        self.connection = None

    def connect(self):
        try:
            self.connection = sqlite3.connect(self.db_name)
            print(f"Connected to database: {self.db_name}")
            result = self.connection.execute(self.query)
            return result.fetchall()
        except sqlite3.Error as e:
            print(f"Error connecting to database: {e}")

    def close(self):
        if self.connection:
            self.connection.close()
            print("Database connection closed.")

def calculate_grwoth_salary(base_salary, rate):
    years = 0
    while years < no_of_years:
        years += 1
        yield (base_salary*12) * ((1 + rate) ** years)

def display_results (name,gross_salary, growth_rate, years):
    print(f"Employee: {name}, Gross Salary: {gross_salary:.2f}, Growth Rate: {growth_rate*100:.2f}%, Years: {years}")

sql_connection = SqlConnection(db_name, select_query)
results = sql_connection.connect()
sql_connection.close()
if results:
    print("Query results:")
    for row in results:
        yearly_growth = calculate_grwoth_salary(row[4], growth_rate)
        for year, gross_salary in enumerate(yearly_growth, start=1):
            display_results(f"{row[0]} {row[1]}", gross_salary, growth_rate, year)
else:    print("No data found or an error occurred.")
````

````python
# example of single level inheritance to create a new class that inherits from SqlConnection and adds a method to calculate the growth salary
db_name = 'employee_financial_insights.db'
select_query = "SELECT first_name, last_name, email, department, base_salary_monthly,query FROM employees"
growth_rate = 0.05
no_of_years = 5

class GrowthSalary(SqlConnection):
    def __init__(self, db_name, query, growth_rate, no_of_years=5):
        super().__init__(db_name, query)
        self.growth_rate = growth_rate
        self.no_of_years = no_of_years
        
    def calculate_growth_salary(self, base_salary):
        # base_salary is monthly; convert to yearly and apply compound growth
        yearly_base = base_salary * 12
        return yearly_base * (1 + self.growth_rate) ** self.no_of_years
    
growth_salary_calculator = GrowthSalary(db_name, select_query, growth_rate, no_of_years)
for row in growth_salary_calculator.connect():
    growth_salary = growth_salary_calculator.calculate_growth_salary(row[4])
    print(f"Employee: {row[0]} {row[1]}, Base Salary: {row[4]*12:.2f}, Growth Salary after {growth_salary_calculator.no_of_years} years: {growth_salary:.2f}")
growth_salary_calculator.close()
print("Growth salary calculation completed and database connection closed.")
````
