## Types of Functions in Python

### 1. Built-in Functions
```python
print("Hello, World!")
length = len([1, 2, 3]) #output 3
```

### 2. User-defined Functions
```python
def greet(name):
    return f"Hello, {name}!"
```

### 3. Anonymous (Lambda) Functions
```python
add = lambda x, y: x + y
```

### 4. Recursive Functions
```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
```

### 5. Higher-order Functions
```python
def apply_func(func, value):
    return func(value)
```

### 6. Generator Functions
```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
```

## Types of Arguments in Python Functions

### 1. Positional Arguments
```python
def add(a, b):
    return a + b
```

### 2. Keyword Arguments
```python
def greet(name, message):
    return f"{message}, {name}!"
```

### 3. Default Arguments
```python
def power(base, exponent=2):
    return base ** exponent
```

### 4. Variable-length Arguments
```python
def sum_all(*args):
    return sum(args)

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

````python
# simple function to multiply numbers without verification of the input
def multiplication_numbers(*args):
    result = 1
    for num in args:
        # print("type of num is:", type(num))
        result *= num
    return result

output = multiplication_numbers(2, 3, 4,5,6)
print ("The multiplication of the numbers without verification is:", output)
````

````python
# simple function to multiply numbers with verification of the input
def multiplication_numbers(*args : int) -> int:
    result = 1
    for num in args:
        # print("type of num is:", type(num))
        if type(num) is not int:
            raise ValueError("All arguments must be integers")
        result *= num
    return result

output = multiplication_numbers(2,3,4,5,6)
print ("The multiplication of the numbers with verification is:", output)
````

````python
# simple function to multiply numbers with verification of the input using isinstance() function
def multiplication_numbers(*args: int) -> int:
    """
    Multiply any number of integer arguments and return the product.
    Raises:
        ValueError: If any argument is not an integer.
    """
    result = 1
    for num in args:
        #isinstance is a built-in function that checks if the object is an instance of the specified class or a subclass thereof. 
        # it is more flexible than type() because it allows for inheritance, while type() checks for exact type matches.
        if not isinstance(num, int): 
            raise ValueError("All arguments must be integers")
        result *= num
    return result

output = multiplication_numbers(2, 3, 4, 5, 6)
print("The multiplication of the numbers using isinstance() is:", output)
````

````python
from math import prod

def multiplication_numbers(*args: int) -> int:
    if not all(isinstance(num, int) for num in args):
        raise ValueError("All arguments must be integers")
    return prod(args) if args else 1 # prod() returns the product of the numbers in the iterable. If the iterable is empty, it returns 1.

output = multiplication_numbers(2, 3, 4, 5)
print("The multiplication of the numbers using prod() is:", output)
````

````python
# **kwargs arguments are key, value pair arguments that are passed to a function. They allow you to pass a variable number of keyword arguments to a function.
def greet_people(greeting, *args, **kwargs):
    """
    Demonstrates the use of *args and **kwargs.
    
    greeting: a required positional argument
    *args: extra positional arguments (names)
    **kwargs: keyword arguments (like age, city)
    """
    
    print(greeting)

    # Handle positional arguments (*args)
    for person in args:
        print(f"Hello {person}!")

    # Handle keyword arguments (**kwargs)
    for key, value in kwargs.items():
        print(f"{key}: {value}")


# Call the function
greet_people(
    "Welcome everyone!",
    "Abhinav", "Rohit", "Priya",      # *args
    age=30, city="Gurgaon", country="India"  # **kwargs
)
````

````python
import sys

sample_tuple = (1, 2, 3, 4, 5) # tuple is immutable
print("type of sample_tuple:", type(sample_tuple))
print(sys.getsizeof(sample_tuple))

nums_list = [i for i in range(1, 100)]
print("type of nums_list:", type(nums_list))
print(sys.getsizeof(nums_list))

# generator expression is a memory efficient way to generate a sequence of numbers.
# this is also stored as an object in memory and it generates the numbers on the fly when we iterate over it.
# it looks like a tuple but it is not a tuple, it is a generator object because this is mutable and it generates the numbers on the fly when we iterate over it.
nums_checker = (i for i in range(1, 10)) 
print("type of nums_checker:", type(nums_checker))
print(sys.getsizeof(nums_checker))
print(next(nums_checker))
print(next(nums_checker))
print(next(nums_checker))

````

## Functional Programming Helpers (`map`, `filter`, `reduce`)

### map()
```python
numbers = [1, 2, 3]
squared = list(map(lambda x: x ** 2, numbers)) #output [1,4,9]
```

### filter()
filter() function is used to extract elements from an iterable (like a list, tuple or set) that satisfy a given condition. It works by applying a function to each element and keeping only those for which function returns True.
```python
even = list(filter(lambda x: x % 2 == 0, [1, 2, 3, 4])) #output [2,4]
```

### reduce()
reduce() applies a binary function (a function that takes two arguments) cumulatively to the items of an iterable, reducing it to a single value.
In Python 3, reduce lives in the functools module.
```python
from functools import reduce
product = reduce(lambda x, y: x * y, [1, 2, 3, 4]) #output 24
```

````python
# Converters
def c_to_f(c: float) -> float:
    return c * 9/5 + 32

def c_to_k(c: float) -> float:
    return c + 273.15

# Sample input
celsius_values = [-40, 0, 25, 37, 100]

# Using map to convert
fahrenheit_values = list(map(c_to_f, celsius_values))
kelvin_values = list(map(c_to_k, celsius_values))

print("Celsius:    ", celsius_values)
print("Fahrenheit: ", fahrenheit_values)
print("Kelvin:     ", kelvin_values)

# as a list of dictionaries
records = list(map(lambda c: {
    "celsius": c,
    "fahrenheit": round(c * 9/5 + 32, 2),
    "kelvin": round(c + 273.15, 2)
}, celsius_values))

print("\nRecords: as list of dictionaries")
for r in records:
    print(r)

# uisng zip to combine the lists into a list of tuples
f_vals = map(c_to_f, celsius_values)
k_vals = map(c_to_k, celsius_values)

# Zip and format with units/symbols; round for neat display
labeled = [
    (f"{c}°C", f"{round(f, 2)}°F", f"{round(k, 2)}K")
    for c, f, k in zip(celsius_values, f_vals, k_vals)
]

print("\nLabeled Records: as list of tuples with units")
for row in labeled:
    print(row)
````

````python
# filter example to filter out words with length greater than or equal to 5
words = ["apple", "banana", "kiwi", "strawberry", "pear"]

long_words = list(filter(lambda w: len(w) >= 5, words))
print(long_words)
````

````python
# reduce example to calculate the sum of a list of numbers
from functools import reduce

nums = [1, 2, 3, 4, 5]
total = reduce(lambda acc, x: acc + x, nums)
print(total)  # 15
````
