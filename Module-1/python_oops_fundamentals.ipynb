### 4. Polymorphism
🔷 Polymorphism = “one name, many forms.”\
It means:\
The same function/method name can work with different types of objects.\
Different classes can implement the same method, and Python will call the correct one based on the object.

```python
class Bird:
    def sound(self):
        print("Some sound")

class Sparrow(Bird):
    def sound(self):
        print("Chirp")

class Parrot(Bird):
    def sound(self):
        print("Squawk")
```

🔷 Does Python Support Method Overloading?
❌ Strict method overloading (like Java/C++)
NO — Python does NOT support compile‑time method overloading.

🔷 Then how does Python handle “overloading”?
Python uses default parameters and variable-length arguments (*args, **kwargs) to simulate overloading.

```python
class Math:
    def add(self, *nums):
        return sum(nums)

m = Math()
print(m.add(10, 20))        # 30
print(m.add(10, 20, 30))    # 60
```
✔ Works like overloading\
✔ But NOT real compile-time overloading\
✔ It is dynamic / runtime-based

🔷 Runtime Polymorphism (Dynamic Polymorphism) — Supported
Yes, Python fully supports runtime polymorphism.
Example — Method Overriding:
```python
class Vehicle:
    def start(self):
        print("Vehicle starting...")

class Car(Vehicle):
    def start(self):
        print("Car starting...")

v = Vehicle()
c = Car()

v.start()   # Vehicle starting...
c.start()   # Car starting...
```

### 5. Abstraction

- Use ABC as the base class.
- Decorate methods with @abstractmethod to force subclasses to implement them.
- A class with even one abstract method becomes abstract.
- Trying to instantiate an abstract class causes a TypeError.
- Subclasses must implement all abstract methods; otherwise, they also remain abstract.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

````python
# example of single class and function to connect to database, execute query, and calculate salary growth
import sqlite3

db_name = 'employee_financial_insights.db'
select_query = "SELECT first_name, last_name, email, department, base_salary_monthly,query FROM employees"
growth_rate = 0.05
no_of_years = 5

class SqlConnection:
    def __init__(self, db_name, query):
        self.__db_name = db_name #enable encapsulation by making db_name private
        self.__query = query

    def connect(self):
        try:
            with sqlite3.connect(self.__db_name) as connection: #use context manager to ensure connection is closed properly
                print(f"Connected to database: {self.__db_name}")
                result = connection.execute(self.__query)
                return result.fetchall()
        except sqlite3.Error as e:
            print(f"Error connecting to database: {e}")

def calculate_grwoth_salary(base_salary, rate):
    years = 0
    while years < no_of_years:
        years += 1
        yield (base_salary*12) * ((1 + rate) ** years)

def display_results (name,gross_salary, growth_rate, years):
    print(f"Employee: {name}, Gross Salary: {gross_salary:.2f}, Growth Rate: {growth_rate*100:.2f}%, Years: {years}")

sql_connection = SqlConnection(db_name, select_query)
results = sql_connection.connect()
if results:
    print("Query results:")
    for row in results:
        yearly_growth = calculate_grwoth_salary(row[4], growth_rate)
        for year, gross_salary in enumerate(yearly_growth, start=1):
            display_results(f"{row[0]} {row[1]}", gross_salary, growth_rate, year)
else:    print("No data found or an error occurred.")
````

````python
# example of single level inheritance to create a new class that inherits from SqlConnection and adds a method to calculate the growth salary
db_name = 'employee_financial_insights.db'
select_query = "SELECT first_name, last_name, email, department, base_salary_monthly,query FROM employees"
growth_rate = 0.05
no_of_years = 5

class GrowthSalary(SqlConnection):
    def __init__(self, db_name, query, growth_rate, no_of_years=5):
        super().__init__(db_name, query)
        self.__growth_rate = growth_rate
        self.__no_of_years = no_of_years

    def get_no_of_years(self):
        return self.__no_of_years  
    
    def calculate_growth_salary(self, base_salary):
        # base_salary is monthly; convert to yearly and apply compound growth
        yearly_base = base_salary * 12
        return yearly_base * (1 + self.__growth_rate) ** self.__no_of_years
    
growth_salary_calculator = GrowthSalary(db_name, select_query, growth_rate, no_of_years)
for row in growth_salary_calculator.connect():
    growth_salary = growth_salary_calculator.calculate_growth_salary(row[4])
    print(f"Employee: {row[0]} {row[1]}, Base Salary: {row[4]*12:.2f}, Growth Salary after {growth_salary_calculator.get_no_of_years()} years: {growth_salary:.2f}")
````

````python
from abc import ABC, abstractmethod

class Shape(ABC): # abstract class always inherits from ABC

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


class Rectangle(Shape):

    def __init__(self, w, h):
        self.__w = w
        self.__h = h

    def area(self):
        return self.__w * self.__h

    def perimeter(self):
        return 2 * (self.__w + self.__h)


rect = Rectangle(10, 5)
print( "Area of the rectangle is:", rect.area())        # 50
print("Perimeter of the rectangle is:", rect.perimeter())   # 30
````

````python
# Absolutely! Here’s a real‑world polymorphism example using a payment processing scenario—something you’d typically find in e‑commerce systems.
# It demonstrates runtime polymorphism (method overriding) and duck typing (no need for a shared base class as long as the method exists).
from abc import ABC, abstractmethod
from dataclasses import dataclass

# 1) Contract
class PaymentProcessor(ABC):
    @abstractmethod
    def pay(self, amount: float) -> str:
        """Process the payment and return a transaction id / receipt."""
        pass

# 2) Implementations
@dataclass
class StripeProcessor(PaymentProcessor):
    api_key: str

    def pay(self, amount: float) -> str:
        # Imagine calling Stripe's SDK/API here
        print(f"[Stripe] Charging ₹{amount:.2f} using API key {self.api_key[:4]}****")
        return "txn_stripe_98765"

@dataclass
class PayPalProcessor(PaymentProcessor):
    client_id: str

    def pay(self, amount: float) -> str:
        print(f"[PayPal] Charging ₹{amount:.2f} via client {self.client_id[:4]}****")
        return "txn_paypal_12345"

class CashOnDelivery(PaymentProcessor):
    def pay(self, amount: float) -> str:
        print(f"[COD] Order placed. Collect ₹{amount:.2f} on delivery.")
        return "txn_cod_placeholder"

# 3) Polymorphic checkout
def checkout(cart_total: float, processor: PaymentProcessor) -> str:
    print(f"Checking out cart with total: ₹{cart_total:.2f}")
    receipt = processor.pay(cart_total)            # <-- polymorphic call
    print(f"Payment successful. Receipt: {receipt}")
    return receipt

# 4) Usage
checkout(1499.00, StripeProcessor(api_key="sk_live_abc123"))
checkout(1499.00, PayPalProcessor(client_id="client_xyz"))
checkout(1499.00, CashOnDelivery())
````

````python
#Option B: Duck Typing (no inheritance required)
# If your team prefers flexibility and lighter coupling, duck typing works great:
#Why it’s powerful
# This is classic Pythonic polymorphism: “If it has a pay method, it’s a payment processor.”
class RazorpayGateway:
    def pay(self, amount: float) -> str:
        print(f"[Razorpay] Charging ₹{amount:.2f}")
        return "txn_razorpay_00001"

class MockGateway:  # Useful in unit tests
    def pay(self, amount: float) -> str:
        print(f"[Mock] Pretend-charging ₹{amount:.2f}")
        return "txn_mock_test"

def checkout(cart_total: float, gateway) -> str:
    # As long as `gateway` has a .pay(amount) method, we’re good.
    receipt = gateway.pay(cart_total)
    return receipt


checkout(799.0, RazorpayGateway())
checkout(799.0, MockGateway())  # perfect for testing without hitting real 
````

````python
#Add-On: Strategy Pattern with a Simple Registry
# Use a registry to pick payment strategy at runtime from config/user choice:

STRATEGIES = {
    "stripe": lambda: StripeProcessor(api_key="sk_live_abc123"),
    "paypal": lambda: PayPalProcessor(client_id="client_xyz"),
    "cod": lambda: CashOnDelivery(),
}

def get_processor(name: str) -> PaymentProcessor:
    try:
        return STRATEGIES[name]()
    except KeyError:
        raise ValueError(f"Unsupported payment method: {name}")

processor = get_processor("paypal")
checkout(2199.00, processor)
````
