## Customer Order Data Display

```python
# In a notebook cell
import pprint

from app import create_app, db
from models import Customer

app = create_app()  # tables are created inside the factory

with app.app_context():
    # All queries go here
    customers = Customer.query.order_by(Customer.created_at.desc()).all()

print("customer tuples : List of tuples with all customer details")
customer_details_tuples = [(f"{c.first_name} {c.last_name}",c.id,c.product,c.category,c.email,c.phone, c.price) for c in customers]
pprint.pprint(customer_details_tuples)


customers_set = set()
customers_list = []
customers_dict = {}
for c in customers:
    customers_set.add( f"{c.first_name} {c.last_name}")
    customers_list.append(f"{c.first_name} {c.last_name}")
    customers_dict[f"{c.first_name} {c.last_name}"] = c.product

print("customer set : Set returns unique values only")
print(customers_set)

print("customer list : List returns all values including duplicates")
print(customers_list)

print("customer dict : Dictionary with customer name as key and id as value")
print(customers_dict)

```
