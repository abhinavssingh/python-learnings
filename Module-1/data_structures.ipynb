## Python Data Structure Types and Their Properties

### 1. List
- **Mutable**
- **Ordered**
- **Allows duplicates**
- **Indexable**
- **Syntax:**
  ```python
  [1, 2, 3]
  ```

### 2. Tuple
- **Immutable**
- **Ordered**
- **Allows duplicates**
- **Indexable**
- **Syntax:**
  ```python
  (1, 2, 3)
  ```

### 3. Set
- **Mutable**
- **Unordered**
- **No duplicates**
- **Not indexable**
- **Syntax:**
  ```python
  {1, 2, 3}
  ```

### 4. Dictionary
- **Mutable**
- **Ordered (Python 3.7+)**
- **Unique keys**
- **Key-value pairs**
- **Syntax:**
  ```python
  {"a": 1, "b": 2}
  ```

### 5. String
- **Immutable**
- **Ordered**
- **Allows duplicates**
- **Indexable**
- **Syntax:**
  ```python
  "hello"
  ```

| Type       | Mutable | Ordered | Duplicates | Indexable | Syntax Example      |
|------------|---------|---------|------------|-----------|---------------------|
| List       | Yes     | Yes     | Yes        | Yes       | `[1, 2, 3]`         |
| Tuple      | No      | Yes     | Yes        | Yes       | `(1, 2, 3)`         |

## Types of Operators in Python

### 1. Arithmetic Operators
Examples: `+`, `-`, `*`, `/`, `//`, `%`, `**`

```python
a = 10
b = 3
print(a + b)   # 13
print(a ** b)  # 1000
```

### 2. Comparison Operators
Examples: `==`, `!=`, `>`, `<`, `>=`, `<=`

```python
print(a > b)    # True
print(a == b)   # False
```

### 3. Logical Operators
Examples: `and`, `or`, `not`

```python
print(a > 5 and b < 5)  # True
print(not(a == b))      # True
```

### 4. Assignment Operators
Examples: `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`

```python
x = 5
x += 3
print(x)  # 8
```

### 5. Bitwise Operators
Examples: `&`, `|`, `^`, `~`, `<<`, `>>`

```python
print(a & b)   # 2
print(a << 1)  # 20
```

### 6. Membership Operators
Examples: `in`, `not in`

```python
nums = [1, 2, 3]
print(2 in nums)       # True
print(4 not in nums)   # True
```

### 7. Identity Operators
Examples: `is`, `is not`

```python
x = [1, 2]
y = x
z = [1, 2]

print(x is y)      # True
print(x is not z)  # True
```

## Set Operations in Python

### Union
```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a | b)  # {1, 2, 3, 4, 5}
```

### Intersection
```python
print(a & b)  # {3}
```

### Difference
```python
print(a - b)  # {1, 2}
```

### Symmetric Difference
```python
print(a ^ b)  # {1, 2, 4, 5}
```

### Subset / Superset
```python
print({1, 2} <= a)  # True
print(a >= {1, 2})  # True
```

## List Comprehensions

### Syntax
```python
[expression for item in iterable if condition]
```

### Examples

#### Squares
```python
squares = [x**2 for x in range(5)]
# [0, 1, 4, 9, 16]
```

#### Filter evens
```python
evens = [x for x in range(10) if x % 2 == 0]
# [0, 2, 4, 6, 8]
```

#### Flatten matrix
```python
matrix = [[1, 2], [3, 4]]
flat = [num for row in matrix for num in row]
# [1, 2, 3, 4]
```


````python
# creating a list of employees with their details
# creating a tuples of employees with their details
# finally creating a dictionary of employees with their query details

# This script connects to the 'employee_financial_insights.db' SQLite database, retrieves employee details.

import sqlite3
import pprint

from zmq import select
db_path = "employee_financial_insights.db"
con = sqlite3.connect(db_path)
cur = con.cursor()


select_query = "SELECT first_name, last_name, email, department, base_salary_monthly, query FROM employees WHERE base_salary_monthly"

cur.execute(select_query)
results = cur.fetchall()
customers = [] # List to store customer names
employee_details = () # Tuple to store employee details
employee_dict = {} # Dictionary to store employee details with name as key

if results:
    customers.clear()  # Clear the list before adding new names
    for row in results:
        customers.append(row[0] + " " + row[1])  # Add the first_name and last_name from each row
        employee_details = (row[0] + " " + row[1], row[2], row[3], row[4], row[5])  # Create a tuple for each employee's details
        employee_dict[row[0] + " " + row[1]] = row[5]  # Add the tuple to the dictionary with the employee's name as the key
else:
    print("No data found or an error occurred.")

con.close()
print("customers list:", customers)
print("employee_details tuple:")
pprint.pprint(employee_details)
print("employee_dict:")
pprint.pprint(employee_dict)
````

## Analyze feedback:
- Create a set containing keywords for classification
- Create a dictionary to store the classification result
- Use a loop to go through each customer's feedback
    - If a feedback message contains words like 'good' or 'excellent' or 'enjopying', classify it as positive
    - If it contains words like 'bad' or 'poor' or 'struggling', classify it as negative
    - Otherwise, classify it as neutral
- Print the classified feedback

````python
# Analyzing the Query Data

positive_words = {"enjoying", "excellent", "great", "amazing" }
negative_words = {"struggling", "poor", "terrible", "slow", "worst"}

feedback_classification = {}
for employee, query in employee_dict.items():
    if any(word in query.lower() for word in positive_words):
        feedback_classification[employee] = "Positive"
    elif any(word in query.lower() for word in negative_words):
        feedback_classification[employee] = "Negative"
    else:
        feedback_classification[employee] = "Neutral"

print("\nFeedback Classification:")
pprint.pprint(feedback_classification)

# Store unique words from all feedback using a set
unique_words = set()
for feedback in employee_dict.values():
    unique_words.update(feedback.lower().split())  # Add words to set

print("\nUnique Words in Feedback:")
pprint.pprint(unique_words)

postive_feedback_customers = [employee for employee, classification in feedback_classification.items() if classification == "Positive"]
print("\nCustomers with Positive Feedback:")
pprint.pprint(postive_feedback_customers)

negative_feedback_customers = [employee for employee, classification in feedback_classification.items() if classification == "Negative"]
print("\nCustomers with Negative Feedback:")
pprint.pprint(negative_feedback_customers)
````

## Organize and display data
- Print customer names with their feedback classification
- Perform set operations to compare feedback categories
- Find customers who gave feedback (union of both sets)
- Find customers who gave both positive and negative feedback (intersection)
- Find customers who gave only negative feedback (difference)


````python
print("\nFeedback Classification:")
pprint.pprint(feedback_classification)

positive_set = set(postive_feedback_customers)
print("\nCustomers with Positive Feedback (Set):")
pprint.pprint(positive_set)

negative_set = set(negative_feedback_customers)
print("\nCustomers with Negative Feedback (Set):")
pprint.pprint(negative_set)

all_positive_and_negative = positive_set.union(negative_set)
sum_of_positive_and_negative = positive_set & negative_set
print("\nCustomers with either Positive or Negative Feedback:")
pprint.pprint(all_positive_and_negative)
print("\nTotal number of customers with either Positive or Negative Feedback:", len(all_positive_and_negative))

who_are_both = positive_set.intersection(negative_set)
print("\nCustomers with both Positive and Negative Feedback:")
pprint.pprint(who_are_both)

difference_positive_negative = positive_set.difference(negative_set)
print("\nCustomers with Positive Feedback but not Negative Feedback:")
pprint.pprint(difference_positive_negative)
````
